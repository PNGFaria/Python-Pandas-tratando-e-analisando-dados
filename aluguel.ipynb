```python
q1 = valor.quantile(.25)
q3 = valor.quantile(.75)
iiq = q3 - q1
lim_inf = q1 - 1.5 * iiq
lim_sup = q3 + 1.5 * iiq
```

```python
selec = (valor >= lim_inf) & (valor <= lim_sup)
dn = res[selec]
```

```python
sns.boxplot(data = dn['Valor'])
```
